```python
pd.reset_option('display.max_rows')
pd.set_option('display.max_columns', None)
data = data.dropna(axis=1)
data
```

칼럼별로 뭘 뜻하는지 좀 알아보자.

```python
# 칼럼명 리스트
columns = list(data.columns)
grouped_columns = {}

for column in columns:
    # 칼럼들을 "_" 앞에 오는 Prefix 기준으로 분리해보자.
    prefix = column.split('_')[0]
    
    # 딕셔너리에 Prefix 기준으로 정리.
    if prefix not in grouped_columns:
        grouped_columns[prefix] = [column]
    else:
        grouped_columns[prefix].append(column)

for key in grouped_columns.keys():
    print(key)
```

VOD는 Vegetation Optical Depth의 약어로, 채집한 마이크로파 센서 데이터를 기반으로 산출된 식물 광학적 깊이를 나타냅니다. 이는 식물 근지반 상부의 물리적 특성과 관련이 있으며, 위층 밀도와 수분 함유량의 변화를 반영합니다. VOD 데이터는 식생량과 연관이 있으며, 이는 GPP와 관련된 중요한 지표입니다. </br>
</br>
여기서 Ku, C, X는 각각 마이크로파의 주파수 대역을 나타냅니다.</br>
</br>
C-band: C-band는 일반적으로 4~8GHz의 주파수 범위를 가지며, 많은 지구 관측 위성에서 사용됩니다. C-band는 토양 습도, 식생량 및 땅의 변위와 같은 지구 표면 속성을 모니터링하는 데 사용됩니다.</br>
</br>
X-band: X-band는 더 높은 주파수 범위인 대략 8~12GHz를 가지며, 주로 대기 및 식생량 관측에 사용됩니다. X-band는 상대적으로 더 고해상도의 데이터를 제공할 수 있으며, 공중 또는 우주에서 더 높은 해상도를 제공합니다.</br>
</br>
Ku-band: Ku-band는 더 높은 주파수 범위인 대략 12~18GHz를 가지며, 통신 및 기타 무선 응용 프로그램에 주로 사용됩니다. 그러나 지구 관측에도 사용될 수 있으며, 이러한 주파수 대역은 토양 습도 및 식생량과 같은 지구 표면 특성을 관측하는 데 유용합니다.
